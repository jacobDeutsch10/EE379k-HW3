```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lines = []
with open('problem2_VGG_pt_readings.txt',  'r') as f:
    lines = [line.split('\t') for line in f.readlines()]
    print(len(lines))
```

```python
pi_lines = []
with open('problem2_VGG_pt_readings_pi.txt',  'r') as f:
    lines = [line.split('\t') for line in f.readlines()]
    print(len(lines))
```

```python
body = np.array([[float(i) for i in line[:-1]] for line in lines[1:]])
print(body)
```

```python
pi_body = np.array([[float(i) for i in line[:-1]] for line in lines[1:]])
```

```python
plt.plot(np.arange(10000),body[:,0],label="Odroid Power")
plt.plot(np.arange(10000),pi_body[:,0],label="Pi Power")
plt.ylabel('Power(W)')
plt.xlabel('Time')
plt.legend()
plt.title("Power Readings for Odroid vs Pi")
plt.show()
```

```python
plt.plot(np.arange(10000),body[:,1],label="Odroid Temp")
plt.plot(np.arange(10000),pi_body[:,1],label="Pi Temp")
plt.ylabel('Temperature')
plt.xlabel('Time')
plt.legend()
plt.title("Temperature Readings for Odroid vs Pi")
plt.show()
```
