```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("stats_pt_1.json", "r") as f:
    pt_data = json.load(f)
pt_data
```

```python
with open("tf_history_1.json", "r") as f:
    tf_data = json.load(f)
tf_data
```

```python

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Training Accuracy over epoch')
ax.set_ylabel('Training Accuracy')
ax.set_xlabel('training epoch')
ax.plot(tf_data['accuracy'], label='tensorflow')
ax.plot(np.array(pt_data['train_acc'])[::3, 2]/100,label='pytorch')
ax.legend()
fig.savefig('q1_train.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Testing Accuracy over epoch')
ax.set_ylabel('Testing Accuracy')
ax.set_xlabel('epoch')
ax.plot(tf_data['val_accuracy'], label='tensorflow')
ax.plot(np.array(pt_data['test_acc'])[:, 2]/100,label='pytorch')
ax.legend()
fig.savefig('q1_test.png')
```
